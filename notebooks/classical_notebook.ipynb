```python
from sklearn.model_selection import train_test_split

X = df_combined["body"]
y = df_combined["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(f"Train Features Shape: {X_train.shape}")
print(f"Train Labels Shape: {X_test.shape}")
print(f"Test Features Shape: {y_train.shape}")
print(f"Test Labels Shape: {y_test.shape}")
```

```python
import re
from bs4 import BeautifulSoup
def clean_email(text):
    # Remove HTML/CSS
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)

    # Remove email addresses
    text = re.sub(r"\S+@\S+", "", text)

    # Remove special characters (keep letters, numbers, and basic punctuation)
    text = re.sub(r"[^a-zA-Z0-9.!?]+", " ", text)

    # Lowercase
    text = text.lower()

    return text

# Apply to train/test data
X_train_clean = X_train.apply(clean_email)
X_test_clean = X_test.apply(clean_email)
```

```python
def extract_features(text):
    features = {}

    # Number of links
    features["num_links"] = len(re.findall(r"http\S+", text))

    # Number of obfuscated words (e.g., "p@ypal")
    features["num_obfuscated"] = len(re.findall(r"\w+[@\$]\w+", text))

    # Urgency keywords
    urgency_words = ["urgent", "immediately", "verify", "password"]
    features["urgency_score"] = sum(text.count(word) for word in urgency_words)

    return features

# Example usage
train_features = X_train_clean.apply(extract_features)
test_features = X_test_clean.apply(extract_features)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF with n-grams (captures phrases like "click here")
tfidf = TfidfVectorizer(
    max_features=5000,          # Control memory usage
    ngram_range=(1, 2),         # Unigrams and bigrams
    stop_words="english"        # Remove common words
)

# Fit on training data only
X_train_tfidf = tfidf.fit_transform(X_train_clean)
X_test_tfidf = tfidf.transform(X_test_clean)
```

```python
import numpy as np
from scipy.sparse import hstack

# Convert manual features to sparse matrix
train_features_sparse = np.array([list(f.values()) for f in train_features])
test_features_sparse = np.array([list(f.values()) for f in test_features])

# Combine with TF-IDF
X_train_final = hstack([X_train_tfidf, train_features_sparse])
X_test_final = hstack([X_test_tfidf, test_features_sparse])
```

```python
import joblib

# Save vectorizers and preprocessed data
joblib.dump(tfidf, "tfidf_vectorizer.pkl")
joblib.dump((X_train_final, y_train, X_test_final, y_test), "preprocessed_data.pkl")
```

```python
X_sample, _, y_sample, _ = train_test_split(
    X_train_final, y_train,
    train_size=0.20,
    stratify=y_train,
    random_state=42
)
```

```python
print(X_sample.shape)
print(y_sample.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import ComplementNB
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import numpy as np

models = {
    # Logistic Regression
    "Logistic Regression": {
        "model": LogisticRegression(class_weight="balanced", max_iter=1000, solver="liblinear"),
        "params": {
            "C": [0.1, 1, 10],
            "penalty": ["l1", "l2"]
        },
        "search_type": "grid"
    },

    # XGBoost
    "XGBoost": {
        "model": XGBClassifier(eval_metric="logloss", use_label_encoder=False, n_jobs=2, verbose=1),
        "params": {
            "max_depth": [3, 5],
            "learning_rate": [0.05, 0.1],
            "subsample": [0.8, 1.0]
        },
        "search_type": "randomized",
        "n_iter": 10
    },

    # Random Forest
    "Random Forest": {
        "model": RandomForestClassifier(class_weight="balanced", n_jobs=-1),
        "params": {
            "n_estimators": [50, 100, 150],
            "max_depth": [5, 10, None],
            "min_samples_split": [2, 5, 10]
        },
        "search_type": "randomized",
        "n_iter": 15
    },


    # LightGBM
    "LightGBM": {
        "model": LGBMClassifier(class_weight="balanced", n_jobs=2, verbose=1),
        "params": {
            "num_leaves": [31, 63, 127],
            "learning_rate": [0.05, 0.1],
            "min_child_samples": [20, 50]
        },
        "search_type": "randomized",
        "n_iter": 10
    },

    # Decision Tree
    "Decision Tree": {
        "model": DecisionTreeClassifier(class_weight="balanced"),
        "params": {
            "max_depth": [5, 10, None],
            "min_samples_split": [2, 5, 10]
        },
        "search_type": "grid"
    },

    # Naive Bayes
    "Naive Bayes": {
        "model": ComplementNB(),
        "params": {
            "alpha": [0.1, 1.0, 10.0]
        },
        "search_type": "grid"
    }
}
```

```python
import torch
print(torch.cuda.is_available())
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import pandas as pd

results = []

for model_name, config in models.items():
    print(f"\n{'='*30}\nTraining {model_name}\n{'='*30}")

    if config["search_type"] == "grid":
        search = GridSearchCV(
            estimator=config["model"],
            param_grid=config["params"],
            scoring="f1",
            cv=3,
            n_jobs=2,
            verbose=3
        )
    else:
        search = RandomizedSearchCV(
            estimator=config["model"],
            param_distributions=config["params"],
            n_iter=config["n_iter"],
            scoring="f1",
            cv=3,
            n_jobs=2,
            verbose=3,
            random_state=42
        )

    search.fit(X_sample, y_sample)

    # Get best model and predictions
    best_model = search.best_estimator_
    y_pred = best_model.predict(X_test_final)

    # Store results
    results.append({
        "Model": model_name,
        "Best Params": search.best_params_,
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
        "Training Time (s)": search.refit_time_,
        "Search Type": config["search_type"]
    })

# Convert to DataFrame
results_df = pd.DataFrame(results)
```

```python
def classify_email(email_text):
    """
    Classifies an email as 'safe' or 'malicious'.

    Args:
        email_text (str): The raw email text to classify

    Returns:
        str: Classification result ('safe' or 'malicious')
    """
    # Load model and vectorizer
    model = joblib.load("malicious_email_classifier.pkl")
    tfidf_vectorizer = joblib.load("tfidf_vectorizer.pkl")

    # Extract manual features before cleaning
    email_manual_features = extract_features(email_text)
    email_manual_features_array = np.array([[
        email_manual_features['num_links'],
        email_manual_features['num_obfuscated'],
        email_manual_features['urgency_score']
    ]])

    # Clean email for TF-IDF
    email_text_clean = clean_email(email_text)
    email_tfidf = tfidf_vectorizer.transform([email_text_clean])

    # Combine features
    email_features_combined = hstack([email_tfidf, email_manual_features_array])

    # Predict and convert to text label
    prediction = model.predict(email_features_combined)
    result = "malicious" if prediction[0] == 1 else "safe"

    # Print result
    print(f"\nEmail: {email_text}")
    print(f"\nClassified as: {result}\n")

    return result
```
